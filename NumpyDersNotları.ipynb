# ***What is NumPy?***



It is a python library that supports multidimensional arrays and matrices, allows us to perform high-level mathematical operations on these arrays and works fast. It is used very often in data analysis.

```python
import numpy as np
```

# ****The Basics****

```python
a = np.array([1,2,3,4])
print(a)
```

```python
b = np.array([1,2,3,'Hello World'])
print(b)
```

```python
skaler = np.array(91)
print(skaler)
```

```python
vektor = np.array([91,92,93,94,95,96,97,98,99])
print(vektor)
```

```python
matris = np.array([[91,92,93],[94,95,96],[97,98,99]])
print(matris)
```
